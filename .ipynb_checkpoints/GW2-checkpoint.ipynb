```python
import pandas as pd
import matplotlib as plt
```

```python
url = "http://www.gw2shinies.com/api/json/history/19721"
```

```python
ecto_history = pd.read_json(url)
```

```python
ecto_history["timestamp"] = pd.to_datetime(ecto_history["timestamp"])
```

```python
ecto_history
```

```python
ecto_history.plot()
```

```python
pd.scatter_matrix(ecto_history, diagonal="hist", figsize=(10,10))
```

```python
ecto_history.describe()
```

```python
ecto_history.index = pd.DatetimeIndex(ecto_history.timestamp)
ecto_history = ecto_history.drop("timestamp", 1)
ecto_history
```

```python
ecto_history[["sell", "buy"]].plot()
```

```python
ecto_history.corr()
```

```python
ecto_history[["supply", "demand"]].plot()
```

```python
ecto_history.supply.diff().plot()
```

```python
ecto_history['2016-05':'2016-02'][["supply", "demand"]].plot()
```

```python
url_daily = "http://www.gw2shinies.com/api/json/history-daily/19721"
ecto_daily_history = pd.read_json(url_daily)
```

```python
ecto_daily_history["timestamp"] = pd.to_datetime(ecto_daily_history["timestamp"])
ecto_daily_history.index = pd.DatetimeIndex(ecto_daily_history.timestamp)
ecto_daily_history = ecto_daily_history.drop("timestamp", 1)
```

```python
ecto_daily_history[["supply", "demand"]].plot()
```

```python
ecto_daily_history[["sell", "buy"]].plot()
```
